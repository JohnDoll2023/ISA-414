There are several ways of solving this problem. Let's start by rerunning all the relevant code from Lecture 7.

```python
import requests
import re
import pandas as pd 

# getting HTML data
r = requests.get("http://www.espn.com/nfl/superbowl/history/winners")
html_code = r.text

# finding open <td> tag
matches = re.finditer("<td>", html_code)
open_tags = [m.start(0) for m in matches]

# finding close <td> tag
matches = re.finditer("</td>", html_code)
close_tags = [m.start(0) for m in matches]

# removing extra </td> tag
close_tags.pop(0)

# retrieving all relevant data from the HTML code
web_data = []
for i in range(len(open_tags)):
    web_data.append(html_code[open_tags[i] + 4:close_tags[i]])


rows = []

# creating a list of lists, each one having 4 elements
for i in range(4, len(web_data), 4):
    rows.append(web_data[i:i+4])

# creating a data frame
df = pd.DataFrame(rows, columns = web_data[0:4])

df
```

Next, I fix the last 13 rows (for the fourth column) of the **df** data frame. To make my solution easier to understand, I have decided to hardcode the row values.


```python
# loop iterating over all problematic rows. See how I hardcode the values 43 to 55
for i in range(42,55):
    
    # for each problematic row, find the relevant open and close tag
    problematic_value = df.iloc[i,3]
    open_tag  = re.search(">",    problematic_value)
    close_tag = re.search("</a>", problematic_value)

    # fix the problem by removing the open/close tags
    fixed_value = problematic_value[open_tag.start(0) + 1:close_tag.start(0)]
    
    # the old problematic value is now equal to the fixed one
    df.iloc[i,3] = fixed_value
    
# see how the wrong values have now been fixed
df.iloc[42:55, 3]
```
