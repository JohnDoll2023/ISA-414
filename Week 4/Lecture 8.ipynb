#### Slide 13: Make sure you install the *request* module first by running *pip install requests* in the Terminal.

```python
import requests
import re

r = requests.get("http://www.espn.com/nfl/superbowl/history/winners")
html_code = r.text

hits = re.finditer("San Francisco [0-9]", html_code)
positions = [m.start(0) for m in hits]
positions
```

#### slide 15

```python
# First regular expression
hits = re.finditer("San Francisco", html_code)
positions = [m.start(0) for m in hits]
positions
```

```python
# Second regular expression
hits = re.finditer("San\sFrancisco", html_code)
positions = [m.start(0) for m in hits]
positions
```

#### slide 16

```python
hits = re.finditer("[A-Z][a-z][a-z]. [0-9][0-9], [0-9][0-9][0-9][0-9]", html_code)
positions = [m.start(0) for m in hits]
positions
```

#### slide 17

```python
hits = re.finditer("[A-Z][a-z][a-z]. [0-9][0-9], [0-9][0-9][0-9][0-9]", html_code)
matches = [m.group(0) for m in hits]
matches
```

#### slide 18

```python
hits = re.finditer("([A-Z][a-z][a-z]). ([0-9][0-9]), ([0-9][0-9][0-9][0-9])", html_code)
matches_group_1 = [m.group(1) for m in hits]
matches_group_1
```

#### slide 20

```python
hits = re.finditer("[A-Z][a-z]{2}. [0-9]{2}, [0-9]{4}", html_code)
matches = [m.group(0) for m in hits]
matches
```

#### slide 21

```python
hits = re.finditer("[A-Z][a-z]{2}. [0-9]{1,2}, [0-9]{4}", html_code)
matches = [m.group(0) for m in hits]
matches
```

#### slide 23

```python
hits = re.finditer("<td>.+?</td>", html_code)
matches = [m.group(0) for m in hits]
matches
```

#### slide 24

```python
hits = re.finditer("<.+>", string = "<a> b c>")
matches = [m.group(0) for m in hits]
matches
```

```python
hits = re.finditer("<.+?>", string = "<a> b c>")
matches = [m.group(0) for m in hits]
matches
```

#### Slide 25

```python
# first code
hits = re.finditer(".", html_code)
matches = [m.start(0) for m in hits]
matches
```

```python
# second code
hits = re.finditer("\.", html_code)
matches = [m.start(0) for m in hits]
matches
```

```python
r = requests.get("https://miamioh.edu/fsb/")
html_code = r.text

hits = re.finditer("cards-news__headline.+?</p>", html_code)
headlines = [m.group(0) for m in hits]
headlines
```
