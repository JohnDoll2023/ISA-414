#### slide 33

```python
# first query
result = collection.find({"mpg":21})

for document in result:
    print(document)
```

```python
# second query
result = collection.find({"mpg":21, "color":"blue"})

for document in result:
    print(document)
```

#### slide 34

```python
# first query
result = collection.find({"mpg":{"$gt":20}})

for document in result:
    print(document)
```

```python
# second query
result = collection.find({"mpg":{"$gt":20, "$lt":25}})
for document in result:
    print(document)
```

#### slide 35

```python
query      = {"name":"Mazda RX4"}
new_values = {"$set":{"mpg":100}}

collection.update_many(query, new_values)

# print all documents after the update
for result in collection.find():
  print(result)
```

#### slide 36

```python
collection.delete_many({"cyl":6})

# print all documents after the delete
for result in collection.find():
  print(result)
```
